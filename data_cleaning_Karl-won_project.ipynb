##Now, let's merge the datasets by country code and time, since they are similar


```python
result = pd.merge(food_price_data, food_consumption_data, on=['Country Code','Time'], how = "right")
```

```python
result.head(15)
```

We see that there is something weird happening with the merging part. Going into the original data, it looks like the first repetition is wrong, and the next one is the right one (second row of 2017 is the right data for 2017). So in order to get rid of this issue, we created a for loop exploting this pattern for all the rows that are incorrect being odd numbers

```python
for i in range(len(result)-1):
  if (i % 2) == 0:
    result = result.drop([i, i])
  else:
    pass
```

```python
result.head(15)
```

Now we have no repetitions and the right datapoints for each row. We believe this issue happened because of the merging in two variables, but it was necessary to have the right data on the right places. We checked multiple countries for the datapoints they have on each row, and it is exactly the result that we expected by merging them. So now, we can work with this data
